```python
import pandas as pd
import numpy as np
```

```python
master_df = pd.read_csv("../../Data/filtered/master-data.csv", sep = ",", encoding = "Latin-1")
```

```python
master_df["IT_experience_in_years"] = master_df["IT_experience_in_years"].fillna('experience_unknown')
```

```python
def normalize_IT_experience(row):
    exp = row
    if row in ["Less than 1 year", "1 - 2 years","<2", "1 to 2 years", "Less than a year", "0-2 years"]:
        exp="below_2"
    elif row in ["2/5/2013", "2/5/2014", "2 - 5 years", "2 to 3 years", "3 to 4 years","4 to 5 years","3-5 years", "41310", "41310"]:
        exp="2_to_5"
    elif row in ["41435", "41070","6/10/2013","6/10/2014","6 - 10 years","9 to 10 years", "6 to 7 years","8 to 9 years",
 "5 to 6 years", "7 to 8 years", "6-8 years", "9-11 years","10 to 11 years"]:
        exp="6_to_10"
    elif row in ["11","11+ years","20 or more years","11 to 12 years","17 to 18 years","14 to 15 years","16 to 17 years","12 to 13 years","18 to 19 years","15 to 16 years","13 to 14 years","19 to 20 years","18-20 years","21-23 years","12-14 years","24-26 years","15-17 years","30 or more years","27-29 years","More than 50 years"]:
        exp="above_11"
    return exp

def categorise_IT_experience(row):
    exp = row 
    exclusion_list = ["below_2","2_to_5","6_to_10","above_11","experience_unknown"]
    if row not in exclusion_list:
        exp_i = int(float(row))
        if exp_i < 2:
            exp="below_2"
        elif exp_i >= 2 and exp_i <=5:
            exp="2_to_5"
        elif exp_i > 5 and exp_i <=10:
            exp="6_to_10"
        elif exp_i > 10:
            exp="above_11"           
    return exp
```

```python
master_df["IT_experience_in_years"] = master_df["IT_experience_in_years"].apply(normalize_IT_experience)
```

```python
master_df["IT_experience_in_years"] = master_df["IT_experience_in_years"].apply(categorise_IT_experience)
```

```python
master_df.groupby(["year", "IT_experience_in_years"])["IT_experience_in_years"].count()
```

```python
master_df = master_df[master_df.IT_experience_in_years != 'experience_unknown']
```

```python
master_df.groupby(["year", "IT_experience_in_years"])["IT_experience_in_years"].count()
```

```python
master_df.to_csv("../../Data/filtered/master-data.csv", index=False)
```
