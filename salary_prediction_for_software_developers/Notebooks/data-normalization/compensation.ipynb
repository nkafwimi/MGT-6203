```python
master_data["compensation"] = master_data["compensation"].fillna('unknown')
```

```python
def normalize_compensation(row):
    compensation = row
    if row in ["<$20,000", "Less than $20,000", "Less than $10,000", "$10,000 - $20,000"]:
        compensation="below-20000"
    elif row in ["$20,000 - $40,000", "$30,000 - $40,000", "$20,000 - $30,000"]:
        compensation="20000-40000"
    elif row in ["$40,000 - $60,000", "$40,000 - $50,000", "$50,000 - $60,000"]:
        compensation="40000-60000"
    elif row in ["$60,000 - $80,000", "$70,000 - $80,000", "$60,000 - $70,000"]:
        compensation="60000-80000"
    elif row in ["$80,000 - $100,000", "$90,000 - $100,000", "$80,000 - $90,000"]:
        compensation="80000-100000" 
    elif row in ["$100,000 - $120,000", "$100,000 - $110,000", "$110,000 - $120,000"]:
        compensation="100000-120000"
    elif row in ["$120,000 - $140,000", "$130,000 - $140,000", "$120,000 - $130,000"]:
        compensation="120000-140000"
    elif row in ["$140,000 - $160,000", ">$140,000", "$140,000 - $150,000", "$150,000 - $160,000"]:
        compensation="140000-160000"
    elif row in ["More than $160,000", "More than $200,000", "$160,000 - $170,000", "$180,000 - $190,000", "$190,000 - $200,000", "$170,000 - $180,000"]:
        compensation="above-160000"
    elif row in ["Rather not say", "Prefer not to say", "Other (please specify)"]:
        compensation="unknown"
    elif row in ["Unemployed", "Student / Unemployed"]:
        compensation="unemployed"
        
    return compensation

def categorise_compensation(row):
    compensation = row 
    exclusion_list = ['unemployed','unknown','80000-100000','20000-40000','60000-80000','40000-60000',
         '140000-160000',
         '100000-120000',
         'below-20000',
         '120000-140000',
         'above-160000',]
    if row not in exclusion_list:
        compensation_i = int(float(row))
        if compensation_i <= 20000:
            compensation="below-20000"
        elif compensation_i > 20000 and compensation_i <=40000:
            compensation="20000-40000"
        elif compensation_i > 40000 and compensation_i <=60000:
            compensation="40000-60000"
        elif compensation_i > 60000 and compensation_i <=80000:
            compensation="60000_to_80000"
        elif compensation_i > 80000 and compensation_i <=100000:
            compensation="80000-100000"
        elif compensation_i > 100000 and compensation_i <=120000:
            compensation="100000-120000"
        elif compensation_i > 120000 and compensation_i <=140000:
            compensation="12000-140000"
        elif compensation_i > 140000 and compensation_i <=160000:
            compensation="140000-160000"
        elif compensation_i > 160000:
            compensation="above-160000"
            
    return compensation
```

```python
master_data["compensation"] = master_data["compensation"].apply(normalize_compensation)
```

```python
master_data["compensation"] = master_data["compensation"].apply(categorise_compensation)
```

```python
master_data = master_data[master_data.compensation != 'unknown']
master_data = master_data[master_data.compensation !='unemployed']
```
