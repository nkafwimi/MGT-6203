### Example create reasonable amount of dummies from lists of values

We likely want to do some (a lot) of additional data cleaning to a column before running this. Especially notice that for some years the list is not currently correctly expanded to catch each unique element

```python
from collections import Counter

all_years = list(df['year'].unique()) # Get a list of all years

entry_in_all_years_full_list = [] # initiate a list to keep unique techs for each year
entry_in_all_years_complete_list = [] # initiate a list to keep all techs for each year


for year in all_years: 
    all_techs = list(df.loc[df['year'] == year, 'all_techs']) #Make a list of list from dataframe
    all_techs = [l for l in all_techs if type(l) == list] # Keep only entries that are actually lists
    flat = [item for sublist in all_techs for item in sublist] # Flatten lists of list
    unique_values = set(flat) # Get unique entries
    entry_in_all_years_complete_list += flat # Add to list of lists of all entries
    entry_in_all_years_full_list.append(unique_values) # Add to flat list of unique entries
    
# Get the intersection of techs that exists for every year
tech_in_all_years = entry_in_all_years_full_list[0].intersection(*entry_in_all_years_full_list)
# In lists of all te
only_relevant_techs = [tech for tech in entry_in_all_years_complete_list if tech in tech_in_all_years]

count_of_relevant_tech_mentions = dict(zip(Counter(only_relevant_techs).keys(), Counter(only_relevant_techs).values()))
count_of_relevant_tech_mentions

# Set min amount of entries for it to be worth to create column
min_entries = 500
techs_for_dummies = [k for k, v in count_of_relevant_tech_mentions.items() if v >= min_entries]
techs_for_dummies



test_str = techs_for_dummies[0]

# Check if element is in list
def element_in_list(lst, element):
    if type(lst) != list:
        0
    elif len([i for i in lst if i == element]) == 0:
        return 0
    else:
        return 1
# Loop through and add dummy-column for everything we would like to keep

for tech in techs_for_dummies: 
    df[tech] = df['all_techs'].apply(element_in_list, args=(tech, ))

    
df
```
