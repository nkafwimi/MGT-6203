This notebook is used to extractor the necessary data for Salary Prediction from 2016 Stack Overflow Annual Developer Survey 

https://insights.stackoverflow.com/survey 

```python
import pandas as pd
import numpy as np
```

NOTE: Make sure **Data/source/** folder has the survey data downloaded using **Scripts/download_survey_data.py --all** script

### 2016 Stack Overflow User Survey

```python
survey_res = pd.read_csv("../../Data/source/2016.csv", sep=",",header = 0, encoding='Latin-1', )
```

```python
print(f"Number of responses to the survey in 2016: {survey_res.shape[0]}")
```

```python
# preview of the data

survey_res.head()
```

```python
survey_res.columns
```

```python
# Extract the Country 

survey_res_flt = pd.DataFrame()
survey_res_flt["country"] = survey_res["country"]

# Extract Age
survey_res_flt["age_range"] = survey_res["age_range"]

# Extract Age
survey_res_flt["age_mp"] = survey_res["age_midpoint"]

#Extract gender
survey_res_flt["gender"] = survey_res["gender"]

# Extract years of IT/programing experience
survey_res_flt["IT_experience_in_years_range"] = survey_res["experience_range"]

# Extract years of IT/programing experience mid point
survey_res_flt["IT_experience_in_years_mp"] = survey_res["experience_midpoint"]

# Extract occupation 
survey_res_flt["occupation"] = survey_res["occupation"]

# Extract salary range
survey_res_flt["salary_range"] = survey_res["salary_range"]

# Extract salary midpoint
survey_res_flt["salary_midpoint"] = survey_res["salary_midpoint"]

# Extract company size
survey_res_flt["company_size"] = survey_res["company_size_range"]

# Extract tech team size
survey_res_flt["team_size"] = survey_res["team_size_range"]

# Extract remote work details 
survey_res_flt["remotely_work_frequency"] = survey_res["remote"]

# Extract time spent on work 
survey_res_flt["work_hours_hobby"] = survey_res["hobby"]

# Extract programming language/technology 
survey_res_flt["proficient_languages"] = survey_res["tech_do"]
```

```python
# Extract the desktop operating system used 
survey_res_flt["desktop_OS"] = survey_res["desktop_os"]

# Extract job switch frequency 
survey_res_flt["job_searching"] = survey_res["open_to_new_job"]
```
