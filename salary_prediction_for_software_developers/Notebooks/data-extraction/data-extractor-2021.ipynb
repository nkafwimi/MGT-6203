This notebook is used to extractor the necessary data for Salary Prediction from 2021 Stack Overflow Annual Developer Survey 

https://insights.stackoverflow.com/survey 

```python
import pandas as pd
import numpy as np
```

NOTE: Make sure **Data/source/** folder has the survey data downloaded using **Scripts/download_survey_data.py --all** script

### 2021 Stack Overflow User Survey

```python
survey_res = pd.read_csv("../../Data/source/2021.csv", sep=",", encoding='Latin-1', low_memory=False)
```

```python
print(f"Number of responses to the survey in 2021: {survey_res.shape[0]}")
```

```python
survey_res.head()
```

```python
# Extract the Country 
survey_res_flt = pd.DataFrame()
survey_res_flt["country"] = survey_res["Country"]

# Extract student information 
survey_res_flt["student"] = survey_res["MainBranch"]

# Extract employment status
survey_res_flt["employment_status"] = survey_res["Employment"]

# Extract employment status
survey_res_flt["formal_education"] = survey_res["EdLevel"]

# Extract company size ~ Number of developers
survey_res_flt["company_size"] = survey_res["OrgSize"]

# Extract occupation 
survey_res_flt["occupation"] = survey_res["DevType"]

# Extract years of IT/programing experience
survey_res_flt["IT_experience_in_years"] = survey_res["YearsCode"]

# Extract years of IT/programing experience Professional
survey_res_flt["IT_experience_in_years_job"] = survey_res["YearsCodePro"]

# Extract job satisfaction 
survey_res_flt["job_satisfaction"] = survey_res["JobSat"]

# Extract compensation
survey_res_flt["compensation"] = survey_res["CompTotal"]

# Extract programming language/technology 
survey_res_flt["proficient_languages"] = survey_res["LanguageHaveWorkedWith"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["database_worked_on"] = survey_res["DatabaseHaveWorkedWith"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["platform_worked_on"] = survey_res["PlatformHaveWorkedWith"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["framework_worked_on"] = survey_res["WebframeHaveWorkedWith"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract the desktop operating system used 
survey_res_flt["desktop_OS"] = survey_res["OpSys"]

# Extract time spent on work 
survey_res_flt["work_hours"] = survey_res["WorkWeekHrs"]

# Extract Age
survey_res_flt["age"] = survey_res["Age"]

#Extract gender
survey_res_flt["gender"] = survey_res["Gender"]
```

```python
# Extract stackoverflow reputation 
survey_res_flt["stackoverflow_account_holder"] = survey_res["SOAccount"]

# Extract stackoverflow reputation 
survey_res_flt["stackoverflow_participation"] = survey_res["SOVisitFreq"]
```
