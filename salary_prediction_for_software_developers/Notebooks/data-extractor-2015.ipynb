```python
import pandas as pd
import numpy as np
```

NOTE: Make sure **Data/source/** folder has the survey data downloaded using **Scripts/download_survey_data.py --all** script

### 2015 Stack Overflow User Survey

```python
survey_res = pd.read_csv("../Data/source/2015.csv", sep=",",header = 1, encoding='Latin-1', )
```

```python
# Extract the Country 

survey_res_flt = pd.DataFrame()
survey_res_flt["country"] = survey_res["Country"]

# Extract Age
survey_res_flt["age"] = survey_res["Age"]

#Extract gender
survey_res_flt["gender"] = survey_res["Gender"]

# Extract years of IT/programing experience
survey_res_flt["IT_experience_in_years"] = survey_res["Years IT / Programming Experience"]

# Extract occupation 
survey_res_flt["occupation"] = survey_res["Occupation"]

# Extract compensation
survey_res_flt["compensation"] = survey_res["Compensation"]

# Extract industry 
survey_res_flt["industry"] = survey_res["Industry"]

# Extract remote work details 
survey_res_flt["remotely_work_frequency"] = survey_res["Remote Status"]

# Extract time spent on work 
survey_res_flt["work_hours_hobby"] = survey_res["How many hours programming as hobby per week?"]
```

```python
# Extract programming language/technology 

language_col_start = survey_res.columns.get_loc("Current Lang & Tech: Android")

survey_res_flt["proficient_languages"] = (
    survey_res
    .iloc[:, language_col_start:language_col_start+42]
    .apply(lambda x: ",".join(x.astype(str)), axis = 1)
)

survey_res_flt["proficient_languages"] = [[pl for pl in pl_list.split(',') if pl != 'nan'] for pl_list in survey_res_flt['proficient_languages'].values.tolist()]
```

```python
# Extract the desktop operating system used 
survey_res_flt["desktop_OS"] = survey_res["Desktop Operating System"]

# Extract job switch frequency 
survey_res_flt["job_switch_in_12_months"] = survey_res["Changed Jobs in last 12 Months"]
```

```python
# Extract stackoverflow reputation 
survey_res_flt["stackoverflow_usage"] = survey_res["How frequently land on or read Stack Overflow"]
```
