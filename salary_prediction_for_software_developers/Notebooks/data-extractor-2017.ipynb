This notebook is used to extractor the necessary data for Salary Prediction from 2017 Stack Overflow Annual Developer Survey 

https://insights.stackoverflow.com/survey 

```python
import pandas as pd
import numpy as np
```

NOTE: Make sure **Data/source/** folder has the survey data downloaded using **Scripts/download_survey_data.py --all** script

### 2017 Stack Overflow User Survey

```python
survey_res = pd.read_csv("../Data/source/2017.csv", sep=",", encoding='Latin-1', low_memory=False)
```

```python
print(f"Number of responses to the survey in 2017: {survey_res.shape[0]}")
```

```python
survey_res.head()
```

```python
# Extract the Country 
survey_res_flt = pd.DataFrame()
survey_res_flt["country"] = survey_res["Country"]

# Extract student information 
survey_res_flt["student"] = survey_res["Professional"]

# Extract employment status
survey_res_flt["employment_status"] = survey_res["EmploymentStatus"]

# Extract employment status
survey_res_flt["formal_education"] = survey_res["FormalEducation"]

# Extract company size ~ Number of developers
survey_res_flt["company_size"] = survey_res["CompanySize"]

# Extract occupation 
survey_res_flt["occupation"] = survey_res["DeveloperType"]

# Extract years of IT/programing experience
survey_res_flt["IT_experience_in_years"] = survey_res["YearsProgram"]

# Extract years of IT/programing experience in Job
survey_res_flt["IT_experience_in_years_job"] = survey_res["YearsCodedJob"]

# Extract job satisfaction 
survey_res_flt["job_satisfaction"] = survey_res["JobSatisfaction"]

# Extract compensation
survey_res_flt["compensation"] = survey_res["Salary"]

# Extract programming language/technology 
survey_res_flt["proficient_languages"] = survey_res["HaveWorkedLanguage"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["database_worked_on"] = survey_res["HaveWorkedDatabase"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["platform_worked_on"] = survey_res["HaveWorkedPlatform"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)

# Extract programming language/technology 
survey_res_flt["framework_worked_on"] = survey_res["HaveWorkedFramework"].apply(lambda x: x.split(";") if not isinstance(x, float) else x)


# Extract time spent on work 
survey_res_flt["work_hours_week"] = survey_res["HoursPerWeek"]

#Extract gender
survey_res_flt["gender"] = survey_res["Gender"]
```
